```python
# import modules required
from bs4 import BeautifulSoup
import requests 
import openpyxl
```

```python

excel = openpyxl.Workbook()    # save output in excel
sheet = excel.active
sheet.title = 'Top Rated Christmas Movies'
sheet.append(['Movie Rank','Movie Name','Year of Release','Rating']) 

# Using the try and catch block to check for errors if any
try:
    source = requests.get('https://editorial.rottentomatoes.com/guide/best-christmas-movies/')
    source.raise_for_status()    # raises error if any
    
    soup = BeautifulSoup(source.text,'html.parser')
    movies = soup.find('body',class_="body").find('div', class_="articleContentBody").\
                                                 find_all('div', class_="row countdown-item")
    
    for movie in movies:
        
        name = movie.find('h2').a.text
        year = movie.find('h2').find('span', class_="subtle start-year").text.strip('()')
        rating = movie.find('h2').find('span', class_="tMeterScore").text
        rank = movie.find('div', class_="countdown-index-resposive").text.strip('#')
        print(rank, name, year, rating) 
        sheet.append([rank, name, year, rating])
    
except Exception as e:
    print(e)
    
excel.save('Top Rated Christmas Movies.xlsx')
```
